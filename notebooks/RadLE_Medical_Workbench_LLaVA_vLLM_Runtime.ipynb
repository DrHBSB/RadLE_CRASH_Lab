# Data Staging Contract

The RadLE v2 image corpus is treated as a frozen snapshot for Workbench runs. Google Drive remains the human/source archive, but inference should read from local Workbench disk after staging from a private regional GCS bucket in the same region as the Workbench VM.

Expected cloud layout:
- Dataset snapshot: `gs://radle-medical-data-toronto/datasets/radle-v2-frozen-2026-06-29/RadLE v2 Master Data/`
- Results archive: `gs://radle-medical-data-toronto/runs/`

Execution data flow:
1. One-time staging: Standard Colab mounts Drive and syncs `RadLE v2 Master Data` to the frozen GCS snapshot.
2. Workbench setup: this notebook copies the frozen snapshot from GCS to local disk through `RADLE_DATASET_GCS_URI`.
3. Benchmark execution: vLLM reads images from the local `RadLE v2 Master Data` folder.
4. Results preservation: after export, the run folder is synced back to GCS through `RADLE_RESULTS_GCS_URI`.

Do not run inference directly from Google Drive or a Drive FUSE mount.

```python
# ==========================================
# 1. RUNTIME SETUP: REPO CODE ONLY
# ==========================================
import os
import pathlib
import subprocess
import sys


def path_is_writable(path):
    try:
        path.mkdir(parents=True, exist_ok=True)
        probe = path / ".radle_write_probe"
        probe.write_text("ok", encoding="utf-8")
        probe.unlink(missing_ok=True)
        return True
    except Exception:
        return False


def select_runtime_root():
    override = os.environ.get("RADLE_RUNTIME_ROOT")
    if override:
        root = pathlib.Path(override).expanduser()
        root.mkdir(parents=True, exist_ok=True)
        return root

    for candidate in (pathlib.Path("/content"), pathlib.Path.home(), pathlib.Path.cwd()):
        if candidate.exists() and path_is_writable(candidate):
            return candidate

    raise RuntimeError("Set RADLE_RUNTIME_ROOT to a writable runtime path.")


def find_existing_repo_dir():
    override = os.environ.get("RADLE_REPO_DIR")
    if override:
        return pathlib.Path(override).expanduser()

    cwd = pathlib.Path.cwd()
    for candidate in (cwd, cwd.parent):
        if (candidate / "src" / "radle_medical_custom_runtime.py").exists():
            return candidate

    return RUNTIME_ROOT / "RadLE_CRASH_Lab"


TOKEN_FILE_BY_SECRET = {
    "GITHUB_TOKEN": ".github_token",
    "HF_TOKEN": ".hf_token",
    "HUGGING_FACE_HUB_TOKEN": ".hf_token",
}


def get_secret(name, *fallback_names):
    for env_name in (name, *fallback_names):
        value = os.environ.get(env_name)
        if value:
            return value

    for secret_name in (name, *fallback_names):
        token_file_name = TOKEN_FILE_BY_SECRET.get(secret_name)
        if not token_file_name:
            continue
        token_file = pathlib.Path.home() / token_file_name
        if token_file.exists():
            value = token_file.read_text(encoding="utf-8").strip()
            if value:
                return value

    try:
        # pyrefly: ignore[missing-import] - optional Colab runtime module.
        from google.colab import userdata
    except Exception:
        return None

    for secret_name in (name, *fallback_names):
        try:
            value = userdata.get(secret_name)
        except Exception:
            value = None
        if value:
            return value
    return None


RUNTIME_ROOT = select_runtime_root()
REPO_URL = "https://github.com/DrHBSB/RadLE_CRASH_Lab.git"
REPO_BRANCH = os.environ.get("RADLE_REPO_BRANCH", "codex/llava-vllm-runtime").strip() or "codex/llava-vllm-runtime"
REPO_REVISION = os.environ.get("RADLE_REPO_REVISION", "").strip()
EXPECTED_LLAVA_MODEL_NAME = "llava_med_mistral_7b"
EXPECTED_LLAVA_MODEL_ID = "chaoyinshe/llava-med-v1.5-mistral-7b-hf"
EXPECTED_SERVER_ENGINE = "vllm"
REPO_DIR = find_existing_repo_dir()
SRC_DIR = REPO_DIR / "src"
MODULE_PATH = SRC_DIR / "radle_medical_custom_runtime.py"
github_token = get_secret("GITHUB_TOKEN")
if github_token:
    os.environ["GITHUB_TOKEN"] = github_token
    print("GitHub token loaded from environment, Colab Secrets, or ~/.github_token; value hidden.")
else:
    print("No GitHub token loaded. Existing local checkouts may still work, but cloning/pulling this private repo needs GITHUB_TOKEN or ~/.github_token.")


def authenticated_repo_url():
    if not github_token:
        return REPO_URL
    return REPO_URL.replace("https://", f"https://x-access-token:{github_token}@")


def mask_secret(value):
    text = str(value)
    if github_token:
        text = text.replace(github_token, "***")
    return text


def run_command(args, check=True, env=None):
    result = subprocess.run(args, text=True, capture_output=True, env=env)
    stdout = mask_secret(result.stdout)
    stderr = mask_secret(result.stderr)
    if stdout.strip():
        print(stdout)
    if stderr.strip():
        print(stderr)
    if check and result.returncode != 0:
        safe_args = [mask_secret(arg) for arg in args[:3]]
        raise RuntimeError(f"Command failed with exit code {result.returncode}: {safe_args}")
    return result


def checkout_runtime_repo():
    repo_url = authenticated_repo_url()
    if (REPO_DIR / ".git").exists():
        if github_token:
            run_command(["git", "-C", str(REPO_DIR), "remote", "set-url", "origin", repo_url])

        print("Fetching repo branch:", REPO_BRANCH)
        fetch_refspec = f"{REPO_BRANCH}:refs/remotes/origin/{REPO_BRANCH}"
        fetch_result = run_command(["git", "-C", str(REPO_DIR), "fetch", "origin", fetch_refspec], check=False)
        if fetch_result.returncode != 0:
            raise RuntimeError(
                "git fetch failed; the runtime cannot refresh the LLaVA-vLLM branch. "
                "Check GITHUB_TOKEN or the existing checkout credentials, then rerun this setup cell."
            )

        checkout_result = run_command(["git", "-C", str(REPO_DIR), "checkout", REPO_BRANCH], check=False)
        if checkout_result.returncode != 0:
            run_command(["git", "-C", str(REPO_DIR), "checkout", "-b", REPO_BRANCH, f"origin/{REPO_BRANCH}"])

        pull_result = run_command(["git", "-C", str(REPO_DIR), "pull", "--ff-only", "origin", REPO_BRANCH], check=False)
        if pull_result.returncode != 0:
            raise RuntimeError(
                "git pull failed; the runtime checkout may be stale, dirty, or divergent. "
                f"Restart the runtime or remove {REPO_DIR}, then rerun this setup cell."
            )
    elif not MODULE_PATH.exists():
        if REPO_DIR.exists() and any(REPO_DIR.iterdir()):
            raise RuntimeError(
                f"{REPO_DIR} exists but is not a Git checkout and {MODULE_PATH} was not found. "
                "Set RADLE_REPO_DIR to a clean checkout, remove that folder, or restart the runtime."
            )

        if not github_token:
            raise RuntimeError(
                "This private GitHub repo needs GITHUB_TOKEN as an environment variable, "
                "Colab secret, or ~/.github_token so the runtime can clone the medical runtime module."
            )

        run_command(["git", "clone", "--branch", REPO_BRANCH, "--single-branch", repo_url, str(REPO_DIR)])
    else:
        print("Using non-Git local module path; verifying helper source directly:", MODULE_PATH)

    if REPO_REVISION:
        print("Checking out explicit repo revision:", REPO_REVISION)
        run_command(["git", "-C", str(REPO_DIR), "checkout", REPO_REVISION])


def assert_expected_helper_source():
    module_text = MODULE_PATH.read_text(encoding="utf-8")
    required_fragments = [
        f'name="{EXPECTED_LLAVA_MODEL_NAME}"',
        f'model_id="{EXPECTED_LLAVA_MODEL_ID}"',
        f'preferred_engine="{EXPECTED_SERVER_ENGINE}"',
    ]
    missing = [fragment for fragment in required_fragments if fragment not in module_text]
    if missing:
        raise RuntimeError(
            "Medical runtime helper is not the LLaVA-vLLM branch mapping. "
            f"Expected {EXPECTED_LLAVA_MODEL_NAME} -> {EXPECTED_LLAVA_MODEL_ID} via {EXPECTED_SERVER_ENGINE}; "
            f"missing source fragments: {missing}. Remove {REPO_DIR} or set RADLE_REPO_BRANCH={REPO_BRANCH!r}, then rerun."
        )


print("Runtime root:", RUNTIME_ROOT)
print("Repo dir:", REPO_DIR)
print("Repo branch:", REPO_BRANCH)
if REPO_REVISION:
    print("Repo revision override:", REPO_REVISION)

checkout_runtime_repo()

if not MODULE_PATH.exists():
    raise RuntimeError(f"Medical runtime module not found after setup: {MODULE_PATH}")

assert_expected_helper_source()

if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

commit_result = run_command(
    ["git", "-C", str(REPO_DIR), "rev-parse", "--short", "HEAD"],
    check=False,
)
branch_result = run_command(
    ["git", "-C", str(REPO_DIR), "rev-parse", "--abbrev-ref", "HEAD"],
    check=False,
)
if branch_result.returncode == 0:
    print("Using repo branch:", branch_result.stdout.strip())
if commit_result.returncode == 0:
    print("Using repo commit:", commit_result.stdout.strip())
print("Verified helper source mapping:", EXPECTED_LLAVA_MODEL_NAME, "->", EXPECTED_LLAVA_MODEL_ID, "via", EXPECTED_SERVER_ENGINE)
```

```python
# ==========================================
# 2. PYTHON DEPENDENCIES + CACHE ROUTING
# ==========================================
# Set INSTALL_SERVER_PACKAGES=False if your custom runtime image already has vLLM.
INSTALL_SERVER_PACKAGES = True
SERVER_ENGINE = "vllm"

# The plain vLLM wheel can resolve to a CUDA-13 build on Colab and fail with
# ImportError: libcudart.so.13. Use the explicit CUDA 12.9 wheel for NVIDIA Colab.
VLLM_VERSION = "0.23.0"
VLLM_CUDA_VARIANT = "cu129"
VLLM_WHEEL_URL = (
    f"https://github.com/vllm-project/vllm/releases/download/v{VLLM_VERSION}/"
    f"vllm-{VLLM_VERSION}%2B{VLLM_CUDA_VARIANT}-cp38-abi3-manylinux_2_28_x86_64.whl"
)
TORCH_EXTRA_INDEX_URL = f"https://download.pytorch.org/whl/{VLLM_CUDA_VARIANT}"

# Workbench runtimes do not always have /content. Keep /content as the Colab
# default, but let RADLE_RUNTIME_ROOT or RADLE_RUNTIME_CACHE_ROOT redirect caches.
ssd_mount = pathlib.Path("/mnt/disks/models_ssd")
if ssd_mount.exists() and os.path.ismount(str(ssd_mount)):
    default_cache_root = ssd_mount
else:
    default_cache_root = RUNTIME_ROOT / "radle_runtime_cache"

RUNTIME_CACHE_ROOT = pathlib.Path(
    os.environ.get("RADLE_RUNTIME_CACHE_ROOT", str(default_cache_root))
).expanduser()
cache_paths = {
    "HF_HOME": RUNTIME_CACHE_ROOT / "huggingface",
    "HUGGINGFACE_HUB_CACHE": RUNTIME_CACHE_ROOT / "huggingface" / "hub",
    "TRANSFORMERS_CACHE": RUNTIME_CACHE_ROOT / "huggingface" / "transformers",
    "PIP_CACHE_DIR": RUNTIME_CACHE_ROOT / "pip",
    "VLLM_CACHE_ROOT": RUNTIME_CACHE_ROOT / "vllm",
    "XDG_CACHE_HOME": RUNTIME_CACHE_ROOT / "xdg",
}
for key, path in cache_paths.items():
    path.mkdir(parents=True, exist_ok=True)
    os.environ[key] = str(path)

print("Cache root:", RUNTIME_CACHE_ROOT)
df_targets = ["/", str(RUNTIME_ROOT)]
if pathlib.Path("/content").exists():
    df_targets.append("/content")
df_targets = list(dict.fromkeys(df_targets))
run_command(["df", "-h", *df_targets], check=False)
run_command(["nvidia-smi"], check=False)

base_packages = [
    "openai",
    "pandas",
    "google-cloud-aiplatform",
    "huggingface_hub",
]
# Do not force --upgrade here. Colab custom runtimes ship pinned packages.
# The medical side path uses local OpenAI-compatible serving; native Anthropic
# and Gemini SDK imports are stubbed by radle_medical_custom_runtime when absent.
run_command([
    sys.executable,
    "-m",
    "pip",
    "install",
    "-q",
    "--cache-dir",
    os.environ["PIP_CACHE_DIR"],
    *base_packages,
])

# Workbench images may ship a scikit-learn wheel built against a different
# NumPy ABI. Transformers treats sklearn as optional, but if the broken wheel
# is present, vLLM's transformers import can fail before model loading.
# RadLE/vLLM do not need sklearn for this path, so remove it from the runtime.
run_command([
    sys.executable,
    "-m",
    "pip",
    "uninstall",
    "-y",
    "scikit-learn",
    "sklearn",
], check=False)

if INSTALL_SERVER_PACKAGES:
    run_command([sys.executable, "-m", "pip", "uninstall", "-y", "vllm"], check=False)
    run_command([
        sys.executable,
        "-m",
        "pip",
        "install",
        "-q",
        "--cache-dir",
        os.environ["PIP_CACHE_DIR"],
        VLLM_WHEEL_URL,
        "--extra-index-url",
        TORCH_EXTRA_INDEX_URL,
    ])
    run_command([
        sys.executable,
        "-c",
        (
            "from vllm.platforms import current_platform; "
            "import vllm; "
            "print('vLLM import OK:', vllm.__version__, vllm.__file__, current_platform)"
        ),
    ])
    flash_attn_probe = run_command([
        sys.executable,
        "-c",
        "\n".join([
            "import importlib.util",
            "flash = importlib.util.find_spec('flash_attn')",
            "rotary = None",
            "if flash is not None:",
            "    try:",
            "        rotary = importlib.util.find_spec('flash_attn.ops.triton.rotary')",
            "    except Exception as exc:",
            "        print('flash_attn rotary probe error:', type(exc).__name__, exc)",
            "print('flash_attn spec:', flash)",
            "print('flash_attn.ops.triton.rotary spec:', rotary)",
            "raise SystemExit(0 if flash is None or rotary is not None else 1)",
        ]),
    ], check=False)
    if flash_attn_probe.returncode != 0:
        print(
            "Removing incompatible flash-attn package so vLLM can use its "
            "rotary fallback instead of importing flash_attn.ops.triton.rotary."
        )
        run_command([
            sys.executable,
            "-m",
            "pip",
            "uninstall",
            "-y",
            "flash-attn",
            "flash_attn",
        ], check=False)
        run_command([
            sys.executable,
            "-c",
            "import importlib.util; print('flash_attn spec after cleanup:', importlib.util.find_spec('flash_attn'))",
        ], check=False)

print("Dependency install finished. Continue unless a later import or server cell raises an error.")
```

```python
# ==========================================
# 4. MODEL, SERVER, AND RUN CONFIG
# ==========================================
import os
from pathlib import Path

# Full-run defaults for the isolated LLaVA-vLLM Workbench route.
# For another shakedown, set TEST_LIMIT to a small integer and change RUN_LABEL_BASE.
# When switching models, stop the server or restart the runtime; RUN_ID is model-scoped.
SELECTED_MODEL_NAME = "llava_med_mistral_7b"

TEST_LIMIT = None
RUN_LABEL_BASE = "medical_full_200_cases"
RUN_ID = f"{SELECTED_MODEL_NAME}_{RUN_LABEL_BASE}"
RESUME = True
MAX_OUTPUT_TOKENS = 2048
EXPECTED_FULL_CASES = 200
EXPECTED_FULL_IMAGES = 263

IS_STANDARD_COLAB = medical_runtime.is_standard_colab()
DEFAULT_DRIVE_DATASET_ROOT = Path(
    "/content/drive/MyDrive/CRASH Lab/RaDLE/CONFIDENTIAL/RadLE v2 Dataset"
)
WORKBENCH_DATASET_ROOT = Path(
    os.environ.get(
        "RADLE_LOCAL_DATASET_ROOT",
        str(Path.home() / "radle_dataset" / "RadLE v2 Dataset"),
    )
).expanduser()

DATASET_ROOT_OVERRIDE = os.environ.get("DATASET_ROOT_OVERRIDE") or os.environ.get("RADLE_DATASET_ROOT", "")
RADLE_DATASET_GCS_URI = os.environ.get(
    "RADLE_DATASET_GCS_URI",
    "gs://radle-medical-data-toronto/datasets/radle-v2-frozen-2026-06-29/RadLE v2 Master Data",
)
os.environ["RADLE_DATASET_GCS_URI"] = RADLE_DATASET_GCS_URI


def has_master_data(path):
    return (Path(path) / "RadLE v2 Master Data").exists()


def maybe_mount_standard_colab_drive():
    if DEFAULT_DRIVE_DATASET_ROOT.exists():
        return
    try:
        # pyrefly: ignore[missing-import] - optional Colab runtime module.
        from google.colab import drive
    except Exception:
        return
    print("Mounting Google Drive for the default RadLE dataset path...")
    drive.mount("/content/drive")


def copy_dataset_from_gcs(dataset_root):
    dataset_root = Path(dataset_root).expanduser()
    if has_master_data(dataset_root):
        print("Dataset already present:", dataset_root)
        return

    source = RADLE_DATASET_GCS_URI.rstrip("/")
    if not source.endswith("/RadLE v2 Master Data"):
        source = f"{source}/RadLE v2 Master Data"

    dataset_root.mkdir(parents=True, exist_ok=True)
    print("Copying RadLE master data from private GCS into:", dataset_root)
    run_command(["gcloud", "storage", "cp", "-r", source, str(dataset_root)])

    if not has_master_data(dataset_root):
        raise RuntimeError(f"GCS copy finished but RadLE v2 Master Data was not found under {dataset_root}")


if (not IS_STANDARD_COLAB) and DATASET_ROOT_OVERRIDE.startswith("/content/drive"):
    raise RuntimeError(
        "DATASET_ROOT_OVERRIDE points to /content/drive, but Workbench/Colab Enterprise "
        "should not rely on google.colab.drive.mount(). Use a local copied folder or GCS."
    )

if IS_STANDARD_COLAB:
    maybe_mount_standard_colab_drive()
else:
    copy_dataset_from_gcs(WORKBENCH_DATASET_ROOT)

_dataset_candidates = []
if DATASET_ROOT_OVERRIDE:
    _dataset_candidates.append(Path(DATASET_ROOT_OVERRIDE).expanduser())
if IS_STANDARD_COLAB:
    _dataset_candidates.append(DEFAULT_DRIVE_DATASET_ROOT)
else:
    _dataset_candidates.append(WORKBENCH_DATASET_ROOT)

seen_candidates = set()
for _candidate in _dataset_candidates:
    candidate_key = str(_candidate)
    if candidate_key in seen_candidates:
        continue
    seen_candidates.add(candidate_key)
    if has_master_data(_candidate):
        dataset_root = _candidate
        break
else:
    checked = "\n".join(f"- {candidate}" for candidate in _dataset_candidates)
    raise RuntimeError(
        "Could not find the RadLE dataset root. Expected one candidate to contain "
        "'RadLE v2 Master Data'.\n"
        f"RADLE_DATASET_GCS_URI={RADLE_DATASET_GCS_URI}\n"
        f"Checked:\n{checked}"
    )

run_paths = medical_runtime.build_medical_run_paths(
    dataset_root,
    model_name=SELECTED_MODEL_NAME,
    run_label=RUN_LABEL_BASE,
    run_id=RUN_ID,
)

master_images_folder = run_paths["master_images_folder"]
raw_results_csv = run_paths["raw_results_csv"]
raw_backup_dir = run_paths["raw_backup_dir"]
scorer_csv = run_paths["scorer_view_csv"]

image_index_for_expected_cases = radle_benchmark.build_image_index(master_images_folder)
all_case_ids = sorted(image_index_for_expected_cases.keys(), key=radle_benchmark.numeric_case_sort_key)
expected_case_ids = all_case_ids if TEST_LIMIT is None else all_case_ids[:TEST_LIMIT]
expected_case_count = len(expected_case_ids)
expected_image_count = sum(len(image_index_for_expected_cases[case_id]) for case_id in expected_case_ids)
multi_image_cases = {
    case_id: [Path(path).name for path in image_index_for_expected_cases[case_id]]
    for case_id in expected_case_ids
    if len(image_index_for_expected_cases[case_id]) > 1
}

if not expected_case_ids:
    raise RuntimeError(f"No benchmark cases found under {master_images_folder}")
if TEST_LIMIT is None and expected_case_count != EXPECTED_FULL_CASES:
    raise RuntimeError(
        f"Full run expected {EXPECTED_FULL_CASES} cases, but found {expected_case_count}. "
        "Check dataset staging before launching the full run."
    )
if TEST_LIMIT is None and expected_image_count != EXPECTED_FULL_IMAGES:
    raise RuntimeError(
        f"Full run expected {EXPECTED_FULL_IMAGES} image files, but found {expected_image_count}. "
        "Check dataset staging before launching the full run."
    )

model_runtime = medical_runtime.get_model(SELECTED_MODEL_NAME)
if SELECTED_MODEL_NAME != EXPECTED_LLAVA_MODEL_NAME:
    raise RuntimeError(
        f"This notebook is locked to {EXPECTED_LLAVA_MODEL_NAME}; got {SELECTED_MODEL_NAME}."
    )
if (
    model_runtime.model_id != EXPECTED_LLAVA_MODEL_ID
    or model_runtime.preferred_engine != EXPECTED_SERVER_ENGINE
):
    raise RuntimeError(
        "Selected runtime mapping is stale or wrong: "
        f"{model_runtime.model_id=} {model_runtime.preferred_engine=}. "
        f"Expected {EXPECTED_LLAVA_MODEL_ID} via {EXPECTED_SERVER_ENGINE}."
    )
print(
    "Selected runtime mapping:",
    SELECTED_MODEL_NAME,
    "->",
    model_runtime.model_id,
    "via",
    model_runtime.preferred_engine,
)
active_model = [model_runtime.benchmark_config()]
REQUEST_EXTRA_BODY = active_model[0].get("extra") or None
# min_tokens=16 ONLY: image cases EOS immediately without it (completion_tokens=1,
# empty); text-only cases generate fine. logit_bias/bad_words were harmful and are
# gone. See the runtime registry comment for the full evidence chain.
EXPECTED_REQUEST_EXTRA_BODY = {"min_tokens": 16}
if REQUEST_EXTRA_BODY != EXPECTED_REQUEST_EXTRA_BODY:
    raise RuntimeError(f"Unexpected LLaVA-vLLM request extra body: {REQUEST_EXTRA_BODY!r}")

# Local server config. Use START_SERVER=False if you already started an OpenAI-compatible endpoint.
START_SERVER = True
HOST = "127.0.0.1"
PORT = 8000
BASE_URL = f"http://{HOST}:{PORT}/v1"
TENSOR_PARALLEL_SIZE = 1
GPU_MEMORY_UTILIZATION = model_runtime.default_gpu_memory_utilization
MAX_MODEL_LEN = 8192
SERVER_TIMEOUT_SECONDS = 900
MODEL_DTYPE = "float16"
LLAVA_CHAT_TEMPLATE_PATH = REPO_DIR / "src" / "templates" / "llava_med_mistral_vllm_chat_template.jinja"
if not LLAVA_CHAT_TEMPLATE_PATH.exists():
    raise RuntimeError(f"Missing LLaVA-vLLM chat template: {LLAVA_CHAT_TEMPLATE_PATH}")
EXTRA_SERVER_ARGS = [
    "--dtype",
    MODEL_DTYPE,
    "--limit-mm-per-prompt",
    "{\"image\": 5}",
    "--chat-template",
    str(LLAVA_CHAT_TEMPLATE_PATH),
    "--chat-template-content-format",
    "openai",
    "--generation-config",
    "vllm",
]

repair_output_csv = run_paths["repair_results_csv"]
repair_call_log_csv = run_paths["repair_call_log_csv"]
repair_plan_csv = run_paths["repair_plan_csv"]
repair_backup_dir = run_paths["repair_backup_dir"]
final_results_csv = run_paths["final_results_csv"]
final_manifest_json = run_paths["final_manifest_json"]
public_release_dir = run_paths["public_release_dir"]

print("Standard Colab:", IS_STANDARD_COLAB)
print("Selected model:", SELECTED_MODEL_NAME)
print("Run label base:", RUN_LABEL_BASE)
print("Model-scoped run ID:", run_paths["run_id"])
print("Test limit:", TEST_LIMIT if TEST_LIMIT is not None else "full")
print("Expected cases this run:", expected_case_count)
print("Expected image files this run:", expected_image_count)
print("Multi-image grouped cases:", len(multi_image_cases))
if multi_image_cases:
    print("Multi-image examples:", dict(list(multi_image_cases.items())[:5]))
print("Dataset GCS URI:", RADLE_DATASET_GCS_URI)
print("Dataset root:", dataset_root)
print("Master images folder:", master_images_folder)
print("Run folder:", run_paths["run_root"])
print("Raw results CSV:", raw_results_csv)
print("Max output tokens:", MAX_OUTPUT_TOKENS)
print("Model dtype:", MODEL_DTYPE)
print("Max model length:", MAX_MODEL_LEN)
print("Extra server args:", EXTRA_SERVER_ARGS)
print("Request extra body:", REQUEST_EXTRA_BODY)
print("Endpoint:", BASE_URL)
```

```python
# ==========================================
# 6. START OR CONNECT TO LOCAL OPENAI-COMPATIBLE SERVER
# ==========================================
previous_server_process = globals().get("server_process")
if START_SERVER and previous_server_process is not None and previous_server_process.poll() is None:
    print("Stopping previous server process before restart...")
    medical_runtime.stop_model_server(previous_server_process)

server_process = None
server_log_path = str(RUNTIME_ROOT / f"{SELECTED_MODEL_NAME}_{SERVER_ENGINE}_server.log")


def cleanup_partial_flash_attn_before_vllm():
    if SERVER_ENGINE != "vllm":
        return
    print("Preflight flash_attn rotary probe before server start...")
    flash_attn_probe = run_command([
        sys.executable,
        "-c",
        "\n".join([
            "import importlib.util",
            "flash = importlib.util.find_spec('flash_attn')",
            "rotary = None",
            "if flash is not None:",
            "    try:",
            "        rotary = importlib.util.find_spec('flash_attn.ops.triton.rotary')",
            "    except Exception as exc:",
            "        print('flash_attn rotary probe error:', type(exc).__name__, exc)",
            "print('flash_attn spec:', flash)",
            "print('flash_attn.ops.triton.rotary spec:', rotary)",
            "raise SystemExit(0 if flash is None or rotary is not None else 1)",
        ]),
    ], check=False)
    if flash_attn_probe.returncode != 0:
        print(
            "Removing incompatible flash-attn package before server start so vLLM can use "
            "its rotary fallback instead of importing flash_attn.ops.triton.rotary."
        )
        run_command([
            sys.executable,
            "-m",
            "pip",
            "uninstall",
            "-y",
            "flash-attn",
            "flash_attn",
        ], check=False)
        run_command([
            sys.executable,
            "-c",
            "import importlib.util; print('flash_attn spec after server preflight cleanup:', importlib.util.find_spec('flash_attn'))",
        ], check=False)


def patch_vllm_rotary_flash_attn_fallback():
    if SERVER_ENGINE != "vllm":
        return
    print("Preflight vLLM rotary optional flash_attn fallback patch...")
    patch_code = r"""
import importlib.util
from pathlib import Path

spec = importlib.util.find_spec("vllm.model_executor.layers.rotary_embedding.common")
if spec is None or spec.origin is None:
    raise SystemExit("Could not locate vLLM rotary common.py")

path = Path(spec.origin)
text = path.read_text(encoding="utf-8")
needle = "\n".join([
    "        self.apply_rotary_emb_flash_attn = None",
    "        if not current_platform.is_cpu() and find_spec(\"flash_attn\") is not None:",
    "            from flash_attn.ops.triton.rotary import apply_rotary",
    "",
    "            self.apply_rotary_emb_flash_attn = apply_rotary",
    "",
])
replacement = "\n".join([
    "        self.apply_rotary_emb_flash_attn = None",
    "        if not current_platform.is_cpu() and find_spec(\"flash_attn\") is not None:",
    "            try:",
    "                from flash_attn.ops.triton.rotary import apply_rotary",
    "            except ModuleNotFoundError as exc:",
    "                if exc.name is None or not exc.name.startswith(\"flash_attn\"):",
    "                    raise",
    "                apply_rotary = None",
    "",
    "            if apply_rotary is not None:",
    "                self.apply_rotary_emb_flash_attn = apply_rotary",
    "",
])

if "except ModuleNotFoundError as exc:" in text:
    print("vLLM rotary optional flash_attn patch already present:", path)
elif needle in text:
    path.write_text(text.replace(needle, replacement, 1), encoding="utf-8")
    print("Patched vLLM rotary optional flash_attn fallback:", path)
else:
    raise SystemExit("Expected vLLM rotary flash_attn block not found in " + str(path))

for pyc_path in path.parent.joinpath("__pycache__").glob("common*.pyc"):
    pyc_path.unlink(missing_ok=True)
"""
    run_command([sys.executable, "-c", patch_code], check=True)


cleanup_partial_flash_attn_before_vllm()
patch_vllm_rotary_flash_attn_fallback()

if START_SERVER:
    print(f"Starting {SERVER_ENGINE} server for {SELECTED_MODEL_NAME}...")
    server_process = medical_runtime.start_model_server(
        model_name=SELECTED_MODEL_NAME,
        engine=SERVER_ENGINE,
        host=HOST,
        port=PORT,
        hf_token=hf_token,
        tensor_parallel_size=TENSOR_PARALLEL_SIZE,
        gpu_memory_utilization=GPU_MEMORY_UTILIZATION,
        max_model_len=MAX_MODEL_LEN,
        extra_args=EXTRA_SERVER_ARGS,
        log_path=server_log_path,
    )
    print("Server process started. Log path:", server_log_path)
    print("Model download/load can take several minutes on the first run.")
else:
    print("START_SERVER=False; connecting to existing endpoint.")

try:
    models_payload = medical_runtime.wait_for_openai_server(
        base_url=BASE_URL,
        timeout_seconds=SERVER_TIMEOUT_SECONDS,
        process=server_process if START_SERVER else None,
        log_path=server_log_path,
    )
except Exception:
    print("Server did not become ready. Last server log lines:")
    print(medical_runtime.read_log_tail(server_log_path, lines=120))
    raise

print("Endpoint /models response:")
print(models_payload)

endpoint_client = medical_runtime.make_openai_client(BASE_URL)
```

```python
# ==========================================
# 6.4. TEXT-ONLY vs IMAGE DIAGNOSTIC PROBE (non-writing, no CSV)
# ==========================================
# The image smoke has repeatedly produced a WHITESPACE LOOP: image tokens ARE
# spliced (prompt_tokens jumps with images) but the model emits only spaces and
# newlines to the token cap. Before adding more token-level band-aids (logit_bias
# / bad_words), split the failure mode with a text-only request through the same
# endpoint, template, and generation-config:
#   * If PLAIN TEXT also degenerates -> the LM decode path / chat template /
#     --generation-config is the fault, and it is fixable in this notebook.
#   * If plain text answers coherently but images break it -> the fault is the
#     image pathway of chaoyinshe/llava-med-v1.5-mistral-7b-hf (projector weights
#     from the HF conversion). No logit_bias fixes that; it is a checkpoint
#     decision for the user.
# This cell only PRINTS. It never writes a CSV row and never raises, so the image
# smoke below still runs in the same session.
import json

if "endpoint_client" not in globals() or endpoint_client is None:
    raise RuntimeError(
        "Section 6 did not complete: endpoint_client is not initialized. "
        "Run Section 6 to readiness before this diagnostic probe."
    )


def _diag_probe(label, content, extra_body=None, max_tokens=256):
    kwargs = {
        "model": model_runtime.model_id,
        "messages": [{"role": "user", "content": content}],
        "temperature": radle_benchmark.UNIVERSAL_TEMPERATURE,
        "max_tokens": max_tokens,
    }
    if extra_body:
        kwargs["extra_body"] = extra_body
    try:
        resp = endpoint_client.chat.completions.create(**kwargs)
    except Exception as exc:  # noqa: BLE001 - diagnostic only
        print(f"--- PROBE [{label}] RAISED:", type(exc).__name__, exc)
        return ""
    raw = resp.choices[0].message.content or ""
    usage = getattr(resp, "usage", None)
    ws = sum(raw.count(ch) for ch in [" ", "\n", "\r", "\t"])
    nonws = len(raw) - ws
    print(f"--- PROBE [{label}] ---")
    print("  finish_reason:", getattr(resp.choices[0], "finish_reason", None))
    print("  prompt_tokens:", getattr(usage, "prompt_tokens", None))
    print("  completion_tokens:", getattr(usage, "completion_tokens", None))
    print("  char_len:", len(raw), "| whitespace:", ws, "| non_whitespace:", nonws)
    print("  raw repr:", repr(raw[:600]))
    return raw


# A) plain text, CLEAN params (no whitespace hacks): can the LM answer at all?
_diag_probe(
    "A_text_only_clean",
    [{"type": "text", "text": "In one short sentence, what is pneumonia?"}],
    extra_body=None,
)
# B) plain text, the real RadLE JSON task, CLEAN params: does it emit JSON?
_diag_probe(
    "B_text_only_radle_json",
    [{"type": "text", "text": radle_benchmark.PROMPT
      + "\n\n(No image is provided in this diagnostic. Return the I don't know JSON.)"}],
    extra_body=None,
)
# C) plain text with the SAME extra_body as the image smoke: isolate the hacks.
_diag_probe(
    "C_text_only_with_smoke_extra_body",
    [{"type": "text", "text": "In one short sentence, what is pneumonia?"}],
    extra_body=REQUEST_EXTRA_BODY,
)
print(
    "\nInterpretation:\n"
    "  * A/B degenerate to whitespace too  -> LM/template/generation-config fault (fix here).\n"
    "  * A/B answer but the image smoke below still whitespaces -> image pathway /\n"
    "    checkpoint conversion fault (escalate to the user; stop token-hacking).\n"
    "Diagnostic probe complete. The image smoke gate runs next."
)
```

```python
# ==========================================
# 6.5. NON-WRITING LLaVA-vLLM SMOKE GATE
# ==========================================
import json

if "endpoint_client" not in globals() or endpoint_client is None:
    raise RuntimeError(
        "Section 6 did not complete: endpoint_client is not initialized. "
        "Rerun Section 6 and require a visible /v1/models response before running Section 6.5."
    )
if "models_payload" not in globals() or not models_payload:
    raise RuntimeError(
        "Section 6 did not complete: /v1/models response is missing. "
        "Stop on the Section 6 server traceback/log instead of running the smoke gate."
    )

smoke_case_id = "156" if "156" in image_index_for_expected_cases else None
if smoke_case_id is None:
    smoke_case_id = max(
        expected_case_ids,
        key=lambda case_id: len(image_index_for_expected_cases[case_id]),
    )

smoke_image_count = len(image_index_for_expected_cases[smoke_case_id])
print(f"Smoke case: {smoke_case_id} ({smoke_image_count} images)")
if smoke_image_count < 2:
    raise RuntimeError(
        "Smoke gate expected a multi-image benchmark case; check dataset staging before the full run."
    )

smoke_content = radle_benchmark.build_content_array(
    smoke_case_id,
    image_index_for_expected_cases,
    prompt=radle_benchmark.PROMPT,
)
smoke_request_kwargs = {
    "model": model_runtime.model_id,
    "messages": [{"role": "user", "content": smoke_content}],
    "temperature": radle_benchmark.UNIVERSAL_TEMPERATURE,
    "max_tokens": MAX_OUTPUT_TOKENS,
}
if REQUEST_EXTRA_BODY:
    smoke_request_kwargs["extra_body"] = REQUEST_EXTRA_BODY
print("Smoke request extra_body:", REQUEST_EXTRA_BODY)
smoke_response = endpoint_client.chat.completions.create(**smoke_request_kwargs)

smoke_raw = smoke_response.choices[0].message.content or ""
smoke_text = smoke_raw.strip()
smoke_usage = getattr(smoke_response, "usage", None)
smoke_prompt_tokens = getattr(smoke_usage, "prompt_tokens", None) if smoke_usage else None
smoke_completion_tokens = getattr(smoke_usage, "completion_tokens", None) if smoke_usage else None
smoke_finish_reason = getattr(smoke_response.choices[0], "finish_reason", None)
print("Smoke raw response:", smoke_raw)
print("Smoke raw response repr:", repr(smoke_raw[:4000]))
print("Smoke raw response char length:", len(smoke_raw))
print("Smoke raw response whitespace counts:", {repr(ch): smoke_raw.count(ch) for ch in [" ", "\n", "\r", "\t"]})
print("Smoke finish reason:", smoke_finish_reason)
print("Smoke prompt tokens:", smoke_prompt_tokens)
print("Smoke completion tokens:", smoke_completion_tokens)

if not smoke_text:
    raise RuntimeError("LLaVA-vLLM smoke failed: empty response.")
if smoke_completion_tokens is not None and smoke_completion_tokens <= 1:
    raise RuntimeError(f"LLaVA-vLLM smoke failed: only {smoke_completion_tokens} completion token(s).")

lower_smoke = smoke_text.lower()
if "<image>" in lower_smoke or "image>" in lower_smoke:
    raise RuntimeError("LLaVA-vLLM smoke failed: response echoed an image placeholder.")
if "you are a board-certified diagnostic radiologist" in lower_smoke or "system prompt" in lower_smoke:
    raise RuntimeError("LLaVA-vLLM smoke failed: response appears to echo the prompt.")

smoke_diag, smoke_likert = radle_benchmark.extract_json_safely(smoke_text)
if smoke_diag in {"PARSE_FAILED", "JSON_MISSING_KEY", None, ""}:
    raise RuntimeError(f"LLaVA-vLLM smoke failed: diagnosis JSON was not parseable: {smoke_diag!r}")

smoke_diag_text = str(smoke_diag).strip()
if smoke_diag_text.lower() == "i don't know":
    raise RuntimeError("LLaVA-vLLM smoke failed: smoke diagnosis was I don't know.")
if smoke_diag_text.lower() in {"unknown", "none", "n/a", "not applicable", "diagnosis"}:
    raise RuntimeError(f"LLaVA-vLLM smoke failed: placeholder diagnosis {smoke_diag_text!r}.")

try:
    smoke_likert_int = int(smoke_likert)
except Exception as exc:
    raise RuntimeError(f"LLaVA-vLLM smoke failed: likert_score is not an integer 0..4: {smoke_likert!r}") from exc
if smoke_likert_int not in range(5):
    raise RuntimeError(f"LLaVA-vLLM smoke failed: likert_score out of range: {smoke_likert!r}")

print(json.dumps({
    "smoke_case_id": smoke_case_id,
    "smoke_image_count": smoke_image_count,
    "diagnosis": smoke_diag_text,
    "likert_score": smoke_likert_int,
}, indent=2))
print("Non-writing smoke gate passed. No benchmark CSV row was written by this cell.")
```

```python
# ==========================================
# 8.5. RAW-OUTPUT INSPECTION (non-writing debug)
# ==========================================
# Show the model's raw responses next to the STORED diagnosis/likert and a
# FRESH re-parse of the same raw text with the current parser. This tells us
# exactly why cells are PARSE_FAILED:
#   * If the FRESH re-parse also fails -> the parser cannot handle the real
#     output format (fix the regex in extract_json_safely).
#   * If the FRESH re-parse SUCCEEDS but the stored value is PARSE_FAILED ->
#     the row is stale from an older (bad_words) run and resume kept it; the
#     run must be forced to re-run those cells (or the CSV cleared).
import pandas as pd

_dbg = pd.read_csv(raw_results_csv, dtype={"Master_Case_ID": str})
_name = SELECTED_MODEL_NAME
_raw_col = f"Raw_Response_{_name}"
_diag_col = f"Diagnosis_{_name}"
_likert_col = f"Likert_{_name}"
_extra_col = f"Actual_Request_Extra_{_name}"
_ts_col = f"Timestamp_UTC_{_name}"

_rows = []
for _, r in _dbg.head(20).iterrows():
    raw = r.get(_raw_col, "")
    raw = "" if pd.isna(raw) else str(raw)
    re_diag, re_likert = radle_benchmark.extract_json_safely(raw)
    _rows.append({
        "case": r.get("Master_Case_ID"),
        "stored_diag": r.get(_diag_col),
        "stored_likert": r.get(_likert_col),
        "reparsed_diag": re_diag,
        "reparsed_likert": re_likert,
        "raw_repr": repr(raw[:160]),
    })
pd.set_option("display.max_colwidth", 220)
pd.set_option("display.max_rows", 60)
display(pd.DataFrame(_rows))

# Full-column counts so we can classify the whole file at a glance.
_stored_fail = 0
_reparse_fail = 0
_reparse_ok_but_stored_fail = 0
for _, r in _dbg.iterrows():
    raw = r.get(_raw_col, "")
    raw = "" if pd.isna(raw) else str(raw)
    stored = str(r.get(_diag_col))
    if stored in {"PARSE_FAILED", "JSON_MISSING_KEY", "nan", "None", ""}:
        _stored_fail += 1
    d, _ = radle_benchmark.extract_json_safely(raw)
    if str(d) in {"PARSE_FAILED", "JSON_MISSING_KEY", "nan", "None", ""}:
        _reparse_fail += 1
    elif stored in {"PARSE_FAILED", "JSON_MISSING_KEY", "nan", "None", ""}:
        _reparse_ok_but_stored_fail += 1

print("Rows total:", len(_dbg))
print("Stored diagnosis empty/PARSE_FAILED:", _stored_fail)
print("Re-parse of Raw_Response STILL fails:", _reparse_fail)
print("Stored PARSE_FAILED but re-parse now SUCCEEDS (STALE rows):", _reparse_ok_but_stored_fail)
if _ts_col in _dbg.columns:
    print("Timestamp range:", _dbg[_ts_col].min(), "->", _dbg[_ts_col].max())
if _extra_col in _dbg.columns:
    print("Sample Actual_Request_Extra:", repr(str(_dbg[_extra_col].dropna().iloc[0])[:300]) if _dbg[_extra_col].notna().any() else "(none)")
```

```python
# ==========================================
# 10. PROMOTE PRIVATE FINAL FILE
# ==========================================
from pathlib import Path

repair_confirmed = globals().get("REPAIR_CONFIRMATION", "NO") != "NO"
repair_output_ready = repair_confirmed and Path(repair_output_csv).exists()
private_final_source_csv = repair_output_csv if repair_output_ready else raw_results_csv
private_final_source_label = "repaired" if repair_output_ready else "raw"

ALLOW_PROMOTE_WITH_PENDING_REPAIRS = False
ALLOW_PROMOTE_WITH_INTEGRITY_WARNINGS = False

# Guardrail: check expected rows/cases and audit targets before promotion.
audit_pre_promote = radle_benchmark.audit_benchmark_output(
    raw_csv=private_final_source_csv,
    models=active_model,
    expected_case_ids=expected_case_ids,
    max_output_tokens=MAX_OUTPUT_TOKENS,
)
pending_repairs = len(audit_pre_promote["repair_targets"])
integrity_table = audit_pre_promote["dataset_integrity"]
integrity = dict(zip(integrity_table["metric"], integrity_table["value"]))


def _as_int(value):
    try:
        return int(value)
    except Exception:
        return -1


integrity_problems = []
if _as_int(integrity.get("rows")) != expected_case_count:
    integrity_problems.append(f"rows={integrity.get('rows')} expected={expected_case_count}")
if _as_int(integrity.get("unique_cases")) != expected_case_count:
    integrity_problems.append(f"unique_cases={integrity.get('unique_cases')} expected={expected_case_count}")
for metric in ["duplicate_case_ids", "missing_expected_case_ids", "extra_case_ids"]:
    if str(integrity.get(metric, "none")) != "none":
        integrity_problems.append(f"{metric}={integrity.get(metric)}")

if integrity_problems and not ALLOW_PROMOTE_WITH_INTEGRITY_WARNINGS:
    print("\nWARNING: dataset integrity problems remain. Promotion halted.")
    for problem in integrity_problems:
        print("-", problem)
    print("Fix the run, or manually set ALLOW_PROMOTE_WITH_INTEGRITY_WARNINGS=True to override.")
elif pending_repairs > 0 and not ALLOW_PROMOTE_WITH_PENDING_REPAIRS:
    print(f"\nWARNING: {pending_repairs} case-model cells still need repair. Promotion halted.")
    print("Run a full repair, or manually set ALLOW_PROMOTE_WITH_PENDING_REPAIRS=True to override.")
else:
    final_manifest = radle_benchmark.promote_final_results(
        source_csv=private_final_source_csv,
        final_csv=final_results_csv,
        manifest_json=final_manifest_json,
        run_id=run_paths["run_id"],
        source_label=private_final_source_label,
        metadata={
            "run_label_base": RUN_LABEL_BASE,
            "model_scoped_run_id": run_paths["run_id"],
            "test_limit": TEST_LIMIT if TEST_LIMIT is not None else "full",
            "expected_case_count": expected_case_count,
            "expected_image_file_count": expected_image_count,
            "multi_image_case_count": len(multi_image_cases),
            "max_output_tokens": MAX_OUTPUT_TOKENS,
        },
    )

    print("Private final source:", private_final_source_csv)
    print("Private final CSV:", final_results_csv)
    print("Private final manifest:", final_manifest_json)
    print("Private final SHA256:", final_manifest["sha256"])

# Rebuild the scorer view from the repaired/promoted final so it reflects repairs.
if Path(final_results_csv).exists():
    radle_benchmark.create_scorer_view(final_results_csv, scorer_csv=scorer_csv)
    print("Scorer view rebuilt from repaired final:", scorer_csv)
```

```python
# ==========================================
# 12. SYNC RESULTS TO GCS
# ==========================================
import json
import platform
import shutil
import sys
from datetime import datetime, timezone
from pathlib import Path

RESULTS_GCS_URI = os.environ.get("RADLE_RESULTS_GCS_URI", "").rstrip("/")
if not RESULTS_GCS_URI:
    raise RuntimeError("Set RADLE_RESULTS_GCS_URI to the private GCS runs folder.")

local_run_root = Path(run_paths["run_root"])
remote_run_root = f"{RESULTS_GCS_URI}/{run_paths['run_id']}"

if not local_run_root.exists():
    raise RuntimeError(f"Local run folder does not exist: {local_run_root}")

runtime_artifacts_dir = local_run_root / "runtime_artifacts"
runtime_logs_dir = runtime_artifacts_dir / "logs"
runtime_provenance_dir = runtime_artifacts_dir / "provenance"
runtime_logs_dir.mkdir(parents=True, exist_ok=True)
runtime_provenance_dir.mkdir(parents=True, exist_ok=True)

log_candidates = []
if globals().get("server_log_path"):
    log_candidates.append(Path(server_log_path))
if globals().get("RUNTIME_ROOT"):
    log_candidates.extend(Path(RUNTIME_ROOT).glob(f"{SELECTED_MODEL_NAME}_*_server.log"))

copied_logs = []
seen_log_paths = set()
for log_path in log_candidates:
    log_path = Path(log_path)
    if not log_path.exists() or log_path in seen_log_paths:
        continue
    seen_log_paths.add(log_path)
    destination = runtime_logs_dir / log_path.name
    shutil.copy2(log_path, destination)
    copied_logs.append(str(destination))

commit_full_result = run_command(
    ["git", "-C", str(REPO_DIR), "rev-parse", "HEAD"],
    check=False,
)
repo_commit_full = commit_full_result.stdout.strip() if commit_full_result.returncode == 0 else ""
run_context = {
    "created_utc": datetime.now(timezone.utc).isoformat(),
    "repo_dir": str(REPO_DIR),
    "repo_commit": repo_commit_full,
    "selected_model": SELECTED_MODEL_NAME,
    "run_id": run_paths["run_id"],
    "run_label_base": RUN_LABEL_BASE,
    "model_scoped_run_id": run_paths["run_id"],
    "test_limit": TEST_LIMIT if TEST_LIMIT is not None else "full",
    "expected_case_count": expected_case_count,
    "expected_image_file_count": expected_image_count,
    "multi_image_case_count": len(multi_image_cases),
    "max_output_tokens": MAX_OUTPUT_TOKENS,
    "max_model_len": MAX_MODEL_LEN,
    "model_dtype": MODEL_DTYPE,
    "extra_server_args": EXTRA_SERVER_ARGS,
    "base_url": BASE_URL,
    "dataset_root": str(dataset_root),
    "master_images_folder": str(master_images_folder),
    "dataset_gcs_uri": RADLE_DATASET_GCS_URI,
    "results_gcs_uri": RESULTS_GCS_URI,
    "local_run_root": str(local_run_root),
    "remote_run_root": remote_run_root,
    "python_executable": sys.executable,
    "python_version": sys.version,
    "platform": platform.platform(),
    "copied_server_logs": copied_logs,
}
run_context_json = runtime_provenance_dir / "run_context.json"
run_context_json.write_text(json.dumps(run_context, indent=2), encoding="utf-8")

print("Archived runtime artifacts:", runtime_artifacts_dir)
print("Copied server logs:", copied_logs if copied_logs else "none found")
print("Run context JSON:", run_context_json)

run_command([
    "gcloud",
    "storage",
    "rsync",
    "-r",
    str(local_run_root),
    remote_run_root,
])

print("Synced local run folder:", local_run_root)
print("Synced GCS run folder:", remote_run_root)
```
